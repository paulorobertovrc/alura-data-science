```python
y = moscow
x = datas
n = np.size(moscow)
```

```python
# REGRESSÃO LINEAR
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2) # coeficiente angular
b = np.mean(y) - a*np.mean(x) # coeficiente linear
```

```python
y = a*x+b
```

```python
np.linalg.norm(moscow-y)
```

```python
plt.plot(datas, moscow)
plt.plot(x, y)
plt.plot(41.5, a*41.5+b, 'ro') # preço da maçã em 41,5 dias
plt.plot(100, a*100+b, 'ro') # previsão do preço da maçã em 100 dias
```
